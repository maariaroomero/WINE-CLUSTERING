CORRELATION MATRIX

```python
correlation_matrix = data.corr()
print(correlation_matrix)
```

HEATMAP

```python
lower_left_matrix = np.tril(correlation_matrix)
mask = lower_left_matrix == 0
plt.figure(figsize=(10, 8))
sns.heatmap(lower_left_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, mask=mask)
column_names = data.columns
plt.xticks(np.arange(len(column_names)) + 0.5, column_names, rotation=45, ha='right')
plt.yticks(np.arange(len(column_names)) + 0.5, column_names, rotation=0)
plt.show()
```

SCATTERPLOTS AND HISTOGRAM 

```python
selected_columns = ['Flavanoids', 'Total_Phenols', 'Nonflavanoid_Phenols', 'Malic_Acid', 'Hue', 'Proline', 'OD280', 'Proanthocyanins']
subset_data = data[selected_columns]
sns.pairplot(data=subset_data, kind='reg', diag_kind='None', plot_kws={'line_kws': {'color': 'green'}}, corner = True)
plt.tight_layout()
plt.show()
```
